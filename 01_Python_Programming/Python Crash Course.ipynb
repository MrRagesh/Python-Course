# Number Guessing Game 

## Level 1. Print if the user guessed number is right or not

```python
secret_number = 65

user_input = int(input("Enter your Guess :"))

if secret_number == user_input :
    print("Congredulations! You're guess is right")
else:
    print("Sorry, Your guess is wrong")
```

## Level 2. Print if the user guessed number is right or not but with max of 7 trials

#### 2.1 Using For Loop - end interation number is defined 

```python
secret_number = 56

for i in range(0,7):
    user_input = int(input("Enter the guessing number : "))
    
    if secret_number == user_input:
        print("Conguratlations!, You'r guessing is right")
        break
    else:
        print("Sorry, You'r guessing is wrong")
```

#### 2.2 While Loop - When end number in not defined and the instructions has to executed infinite times

```python
secret_number = 56

while True:
    user_input = int(input("Enter the guessing number : "))
    
    if secret_number == user_input:
        print("Conguratlations!, You'r guessing is right")
        break
    else:
        print("Sorry, You'r guessing is wrong")
```
